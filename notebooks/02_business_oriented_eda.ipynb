```python
#Churn & Contract Type Analysis
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/data_telco_customer_churn.csv')
```

```python
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title('Churn by Contract Type')
plt.show()
```

```python
#Observation:
#Customers with month-to-month contracts have a significantly higher churn rates.

#Insights:
#Lack of Long-term contract may lead to higher churn.

#Recommendation:
#Offer discounts on yearly plans to reduce churn.
```

```python
#Churn vs Tenure Analysis
```

```python
sns.histplot(data=df,x='tenure',hue='Churn',kde=True,bins=30)
plt.title('Churn vs Tenure Analysis')
plt.show()
```

```python
#Observation:
#Where the churn is concentrated?

#Insight:
#The Higher the tenure ,lesser the churn

#Recommendation:
#Focus on retaining customers in the early stages of their tenure to reduce overall churn.
```

```python
sns.boxplot(x='Churn',y='MonthlyCharges',data=df)
plt.title('Churn vs Monthly Charges')
plt.show()
```

```python
#Observation:
#Customers with higher monthly charges are more likely to churn.

#Insights:
#Higher monthly charges may lead to increased churn.

#Recommendation:
#Consider offering tiered pricing or promotional discounts to make the service more affordable and reduce churn.
```

```python
sns.countplot(x='Churn',hue='InternetService',data=df)
plt.title('Churn by Internet Service')
plt.show()  
```

```python
'''Observations:
Customers with Fiber optic internet service have higher churn rates compared to those with DSL or no internet service.

#Insights:
The type of internet service may influence customer churn, with fiber optic users being more likely to churn.

#Recommendations:
Investigate the reasons behind higher churn among fiber optic users and consider improving service quality or offering incentives to retain these customers.'''
```
